```python
import numpy as np
import matplotlib.pyplot as plt

# Create x values
x = np.linspace(0, 2 * np.pi, 100)

# Create two y values
y1 = np.sin(x)
y2 = np.cos(x)

# Plot both curves
plt.plot(x, y1, label='sin(x)', color='blue')
plt.plot(x, y2, label='cos(x)', color='orange')

# Add labels and legend
plt.title("Sine and Cosine Waves")
plt.xlabel("x (radians)")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Coordinates of the points
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

# Plot the points
plt.scatter(x, y, color='red')

# Add labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot of Points")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# First figure
plt.figure(1)
plt.plot(x, y1, label='sin(x)')
plt.title("Sine Wave")
plt.legend()

# Second figure
plt.figure(2)
plt.plot(x, y2, label='cos(x)', color='orange')
plt.title("Cosine Wave")
plt.legend()

# Show both
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title("Stretched X-Axis")
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 4))  # Wider figure
plt.plot(x, y)
plt.title("Stretched X-Axis")
plt.grid(True)
plt.show()
```

```python
plt.xlim(0, 5)  # Zoom into the first half
```

```python
x_stretched = x * 2
plt.plot(x_stretched, y)
```
