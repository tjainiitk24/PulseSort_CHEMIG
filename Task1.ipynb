```python
import pandas as pd
df_index = pd.read_csv('emg_gesture_recording_starter_kit\\index_finger_data.csv')
df_rest = pd.read_csv('emg_gesture_recording_starter_kit\\rest_data.csv')
df_okay = pd.read_csv('emg_gesture_recording_starter_kit\\okay_data.csv')

print(df_index.head(),'\n')
print(df_rest.head(),'\n')
print(df_okay.head(),'\n')
```

```python
import matplotlib.pyplot as plt
x=df_index['timestamp_ms']

plt.figure(1)
plt.scatter(x,df_index['ch1'],label='ch1',color='blue',s=1)
plt.scatter(x,df_index['ch2'],label='ch2',color='green',s=1)
plt.xlabel("Time in milliseconds")
plt.ylabel("Channel Data")
plt.title("Index Data")
plt.legend()
plt.show()
```

```python
plt.figure(2)
plt.scatter(x,df_rest['ch1'],label='ch1',color='blue',s=1)
plt.scatter(x,df_rest['ch2'],label='ch2',color='green',s=1)
plt.xlabel("Time in milliseconds")
plt.ylabel("Channel Data")
plt.title("Rest Data")
plt.legend()
plt.show()
```

```python
plt.figure(3)
#plt.figure(figsize=(20, 10)) - to stretch the canvas
plt.scatter(x,df_okay['ch1'],label='ch1',color='blue',s=1)
plt.scatter(x,df_okay['ch2'],label='ch2',color='green',s=1)
plt.xlabel("Time in milliseconds")
plt.ylabel("Channel Data")
plt.title("Okay Data")
plt.legend()
plt.show()
```
